```python
#what is the maximum & Minumum No. of crimes recorded per area?
```

```python
# Group by 'area' and calculate maximum and minimum number of crimes
result_crimes = data.groupby('area')['no_of_crimes'].agg(['max', 'min']).reset_index()
```

```python
result_crimes
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# we've already calculated result_crimes
# Group by 'area' and calculate maximum and minimum number of crimes
result_crimes = data.groupby('area')['no_of_crimes'].agg(['max', 'min']).reset_index()

# Plotting
plt.figure(figsize=(12, 6))

# Bar plot for maximum number of crimes
sns.barplot(x='area', y='max', data=result_crimes, color='blue', label='Max Crimes')

# Bar plot for minimum number of crimes
sns.barplot(x='area', y='min', data=result_crimes, color='orange', label='Min Crimes')

# Adding labels and title
plt.title('Maximum and Minimum Number of Crimes Recorded per Area')
plt.xlabel('Area')
plt.ylabel('Number of Crimes')

# Display the legend
plt.legend()

# Rotate x-axis labels for better visibility
plt.xticks(rotation=90)

# Show the plot
plt.show()
```
